```python
file = open('price_and_name.csv')
soup = file.read()
soup = soup.lower()
#soup = replace_function(soup)
bs = BeautifulSoup(soup,"html.parser")
tokens = re.findall('\w+', soup)
#Compile Stopwords
sw = nltk.corpus.stopwords.words('english')
newStopWords = ['1', '2', '3', 'w']
sw.extend(newStopWords)
#Remove Stopwords
words_ns =[]
for word in tokens:
    if word not in sw:
        words_ns.append(word)

cloud_words = " "
cloud_words = cloud_words.join(words_ns)

#Plot top 30
nlp_words = nltk.FreqDist(words_ns)
nlp_words.plot(30)

wordcloud = WordCloud(width = 500, height = 500, max_words = 75, background_color = "#1b212b", colormap = 'Blues').generate(cloud_words)

# Display the generated image:
plt.figure(figsize=[8,8])
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```
